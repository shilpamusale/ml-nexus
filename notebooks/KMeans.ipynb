```python
from sklearn.datasets import make_blobs
from mlnexus.clustering._kmeans import MyKMeans
import matplotlib.pyplot as plt
import numpy as np
```

```python
X, y_true = make_blobs(centers=3, n_samples=500, n_features=2, shuffle=True, random_state=42)
```

```python
# Instantiate and fit the model
kmeans = MyKMeans(k=3)
kmeans.fit(X)
labels = kmeans.predict(X)
```

```python
# Plot the results
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(8, 6))
scatter = ax.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', alpha=0.8)

# Plot the centroids
centroids_array = np.array(kmeans.centroids)
ax.scatter(centroids_array[:, 0], centroids_array[:, 1], c='red', s=250, marker='x', label='Centroids')

plt.title('K-Means Clustering from Scratch')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
