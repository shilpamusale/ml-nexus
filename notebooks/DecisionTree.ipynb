```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
from mlnexus.tree._decision_tree import MyDecisionTree
from sklearn.tree import plot_tree
```

```python
def accuracy(y_true: np.ndarray, y_pred: np.ndarray) -> float:
        """Calculates the accuracy of predictions."""
        accuracy = np.sum(y_true == y_pred) / len(y_true)
        return accuracy
```

```python
# Load a sample dataset
data = datasets.load_breast_cancer()
X, y = data.data, data.target
feature_names = data.feature_names

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1234
)

# Instantiate and train the Decision Tree classifier
clf = MyDecisionTree(max_depth=10)
clf.fit(X_train, y_train)

# Make predictions on the test set
y_pred = clf.predict(X_test)
    
# Calculate and print the accuracy
acc = accuracy(y_test, y_pred)
print(f"Accuracy: {acc:.4f}")
```

# Use SKLearn library to use its Decision tree implementation

```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn import datasets
from sklearn.model_selection import train_test_split

# --- Load Data ---
data = datasets.load_breast_cancer()
X, y = data.data, data.target
feature_names = data.feature_names
class_names = data.target_names.tolist() # Get class names as strings

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1234
)

# --- Use scikit-learn's Decision Tree ---
# 1. Create and train a standard sklearn classifier
sklearn_clf = DecisionTreeClassifier(max_depth=3, random_state=1234)
sklearn_clf.fit(X_train, y_train)
accuracy = sklearn_clf.score(X_test, y_test)

print(f"Accuracy: {accuracy:.4f}")
# --- Use the built-in plot_tree function ---
# 2. Plot the tree
plt.figure(figsize=(20,10)) # Set the figure size
plot_tree(
    sklearn_clf,
    filled=True, # Fill nodes with colors
    feature_names=feature_names, # Use feature names
    class_names=class_names, # Use class names
    rounded=True,
    fontsize=10
)
plt.show()
```
